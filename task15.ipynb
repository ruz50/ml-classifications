```python
def Power():
    print(2**3)

Power()
```

```python
def Power2(x, a):
    print(x**a)

Power2(3, 8)
```

```python
Power2(10, 3)
Power2(8, 17)
Power2(13, 13)
```

```python
def Power3(x, a):
    result = x**a
    return result

Power3(2, 5)
```

```python
import numpy as np
import matplotlib.pyplot as plt

x = np.arange(1, 11)
y = [Power3(i, 2) for i in x]

plt.plot(x, y)
plt.xlabel("x")
plt.ylabel("x^2")
plt.title("f(x) = x^2")

# optional log scale
# plt.xscale('log')
# plt.yscale('log')

plt.show()
```

```python
def PlotPower(x, a):
    y = [i**a for i in x]
    
    plt.plot(x, y)
    plt.xlabel("x")
    plt.ylabel(f"x^{a}")
    plt.title(f"Plot of x vs x^{a}")
    
    plt.show()
```

```python
import numpy as np

PlotPower(np.arange(1, 11), 3)
```
